```python
# Initialize Reddit API
reddit = praw.Reddit(
    client_id=REDDIT_CLIENT_ID,
    client_secret=REDDIT_CLIENT_SECRET,
    user_agent=REDDIT_USER_AGENT
)
```

```python
def fetch_user_data(username, max_items=50):
    user = reddit.redditor(username)
    posts = []
    comments = []

    try:
        for submission in user.submissions.new(limit=max_items):
            posts.append({
                'type': 'post',
                'title': submission.title,
                'selftext': submission.selftext,
                'url': submission.url
            })

        for comment in user.comments.new(limit=max_items):
            comments.append({
                'type': 'comment',
                'body': comment.body,
                'link': f"https://reddit.com{comment.permalink}"
            })
    except Exception as e:
        print(f"Error fetching data: {e}")

    return posts, comments
```

```python
reddit_url = input("Enter Reddit profile URL: ").strip()
username = reddit_url.rstrip('/').split("/")[-1]

print(f"Fetching data for: {username}")
posts, comments = fetch_user_data(username)

print(f"Fetched {len(posts)} posts and {len(comments)} comments.")
```

```python
import httpx

def generate_persona_openrouter(posts, comments, api_key):
    text_chunks = []
    citations = []

    for i, item in enumerate(posts + comments):
        content = item.get('title', '') + " " + item.get('selftext', '') + item.get('body', '')
        citation = item.get('url') or item.get('link')
        text_chunks.append(f"{i+1}. {content.strip()[:1000]}")
        citations.append(citation)

    prompt = (
        "You're an expert analyst. Build a user persona using the following Reddit posts and comments:\n\n"
        + "\n\n".join(text_chunks)
        + "\n\nReturn a structured persona with traits like Age, Gender (if possible), Interests, Political views, "
          "Personality traits, Writing style, etc. Include references to the post/comment number used for each trait."
    )

    headers = {
        "Authorization": f"Bearer {api_key}",
        "HTTP-Referer": "https://chat.openai.com",  # Optional but helps prevent blocking
        "Content-Type": "application/json"
    }

    payload = {
        "model": "mistralai/mistral-7b-instruct:free",  # ✅ Replace model name
        "messages": [{"role": "user", "content": prompt}]
    }

    response = httpx.post(
        "https://openrouter.ai/api/v1/chat/completions",
        headers=headers,
        json=payload
    )

    result = response.json()
    print("🔁 Full API response:")
    print(result)
    content = result["choices"][0]["message"]["content"]
    return content, citations
```

```python
import json

def save_persona_to_json(persona_text, citations, filename="user_persona.json"):
    data = {
        "persona": persona_text,
        "citations": citations
    }
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
    print(f"✅ Persona saved to {filename}")
```

```python
def save_persona_to_html(persona_text, citations, filename="user_persona.html"):
    html = f"""
    <html>
    <head>
        <title>User Persona</title>
        <style>
            body {{ font-family: Arial, sans-serif; padding: 20px; }}
            h1 {{ color: #333; }}
            .citations {{ font-size: 0.9em; color: #777; }}
        </style>
    </head>
    <body>
        <h1>User Persona</h1>
        <pre>{persona_text}</pre>
        <h2>Citations</h2>
        <ul>
            {''.join(f"<li>{c}</li>" for c in citations)}
        </ul>
    </body>
    </html>
    """
    with open(filename, "w", encoding="utf-8") as f:
        f.write(html)
    print(f"✅ Persona saved as HTML report to {filename}")
```
